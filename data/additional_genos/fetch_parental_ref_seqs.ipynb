```python
Entrez.email = "sonajeser@gmail.com"  # Always tell NCBI who you are
database = "nucleotide"
geno_dict = {}
for cur_column in range(2, positions.shape[1]):
    cur_maize = positions.columns[cur_column]
    print(f"Processing column {cur_column} out of {positions.shape[1]} ({cur_maize})")
    cur_genos = []
    for cur_row in range(positions.shape[0]):
        # Downloading...
        pos = int(positions.iloc[cur_row, 1])
        print(f"Processed {cur_row/positions.shape[0] * 100}%")
        chr_name = positions.iloc[cur_row, cur_column]
        try:
            stream = Entrez.efetch(db=database, id=chr_name, rettype="fasta",# retmode="text",azaa
                                   seq_start = pos, seq_stop = pos, strand = 2)
            seq = SeqIO.read(stream, "fasta")
            cur_genos.append(str(seq.seq))
            stream.close()
        except Entrez.HTTPError:
            print(f"Fetching position {pos} failed, inserting NA instead")
            cur_genos.append(str(np.nan))
    geno_dict[cur_maize] = cur_genos
    cur_genos = pd.Series(cur_genos)
    print(f"Writing {chr_name} to disk")
    cur_genos.to_csv("./data/additional_genos/" + chr_name + "_genos_str2.csv")
```
